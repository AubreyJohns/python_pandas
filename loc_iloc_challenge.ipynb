# loc and iloc challenge


```python
import pandas as pd
Africa = {
    "Country" : ['Kenya', 'Uganda', 'Tanzania','Somalia','SouthAfrica','Nigeria','Niger'],
    "Capital" : ['Nairobi', 'Kampala', 'Daressaalam','Mogadishu','CapeTown','Abuja','Niamey'],
    "Area" : [205, 189, 250,120,280,360,90],
    "Population" : [49, 32, 50,20,40,80,20]
}
```

```python
Africa_df = pd.DataFrame(Africa)
Africa_df
```

```python
#Saving dataframe into a CSV file

Africa_df.to_csv('africa.csv', index = False)
```

```python
#index_col will make the column specified inside it to be the new row label
Africa_df = pd.read_csv('africa.csv', index_col = ['Country'])
Africa_df
```
