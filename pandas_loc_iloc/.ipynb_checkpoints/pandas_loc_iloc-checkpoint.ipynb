```python
df1 = titanic_df.loc[[0,1,2,3,4,5],['PassengerId', 'Survived', 'Gender']]
df1
```

```python
df2 = titanic_df.loc[[0,2,3,5],['PassengerId', 'Age', 'Parch']]
df2
```

```python
print("Inner - Natural Join")
pd.merge(df1, df2, on='PassengerId', how='inner')
```

```python
print("Outer - Full Outer Join")
pd.merge(df1, df2, on='PassengerId', how='outer')
```

```python
print("Left - Left Outer Join")
pd.merge(df1, df2, on='PassengerId', how='left')
```
