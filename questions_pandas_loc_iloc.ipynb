```python
import pandas as pd

# question 1
Car_Price = {
    'Company': ['Toyota', 'Honda', 'BMV', 'Audi'],
    'Price': [23845, 17995, 135925 , 71400]
}
car_Horsepower = {
    'Company': ['Toyota', 'Honda', 'BMV', 'Audi'],
    'horsepower': [141, 80, 182 , 160]
}
```

```python
Car_Price_pd = pd.DataFrame(Car_Price)
car_Horsepower = pd.DataFrame(car_Horsepower)
```

```python
# inner
print("Inner - Natural Join")
pd.merge(Car_Price_pd, car_Horsepower, on='Company', how='inner')
```

```python
# outer
print("Outer - Full Outer Join")
pd.merge(Car_Price_pd, car_Horsepower, on='Company', how='outer')
```

```python
# left
print("Left - Left Outer Join")
pd.merge(Car_Price_pd, car_Horsepower, on='Company', how='left')
```

```python
# right
print("Right - Right Outer Join")
pd.merge(Car_Price_pd, car_Horsepower, on='Company', how='right')
```
